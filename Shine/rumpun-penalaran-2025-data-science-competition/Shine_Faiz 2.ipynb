```python
import pandas as pd
import numpy as np
import xgboost as xgb  # Mengimpor XGBoost
from sklearn.impute import KNNImputer
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# 1. MEMUAT DATASET
# ==============================================================================
# Menggunakan path file yang Anda tentukan
file_path = "C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/train_dataset.csv"

try:
    df = pd.read_csv(file_path)
    print(f"Dataset dari '{file_path}' berhasil dimuat.")
except FileNotFoundError:
    print(f"File tidak ditemukan di path: '{file_path}'. Pastikan path sudah benar.")
    # Keluar dari skrip jika file tidak ada, karena tidak ada fallback
    exit()

# ==============================================================================
# 2. PERSIAPAN UNTUK ANALISIS
# ==============================================================================
# Identifikasi fitur numerik
numerical_features = df.select_dtypes(include=np.number).columns.tolist()

# Pilih kolom target kategorikal untuk diimputasi (misalnya, 'school_type')
# Anda bisa mengganti ini jika targetnya adalah kolom lain
target_col = 'school_type'
print(f"Kolom target kategorikal untuk diimputasi: '{target_col}'")
print("-" * 50)

# Buat salinan DataFrame untuk setiap metode imputasi
df_mean = df.copy()
df_median = df.copy()
df_knn = df.copy()

# ==============================================================================
# 3. LAKUKAN IMPUTASI PADA FITUR NUMERIK
# ==============================================================================
# Metode 1: Imputasi Mean
df_mean[numerical_features] = df_mean[numerical_features].fillna(df_mean[numerical_features].mean())

# Metode 2: Imputasi Median
df_median[numerical_features] = df_median[numerical_features].fillna(df_median[numerical_features].median())

# Metode 3: Imputasi KNN
knn_imputer = KNNImputer(n_neighbors=5) # Menggunakan 5 tetangga
df_knn[numerical_features] = knn_imputer.fit_transform(df_knn[numerical_features])

# ==============================================================================
# 4. EVALUASI SETIAP METODE IMPUTASI DENGAN XGBOOST
# ==============================================================================
imputation_methods = {
    'Mean': df_mean,
    'Median': df_median,
    'KNN': df_knn
}
results = {}

print(f"Mengevaluasi performa untuk memprediksi '{target_col}' menggunakan XGBoost...")

for method_name, imputed_df in imputation_methods.items():
    # Gunakan hanya baris di mana variabel target diketahui untuk evaluasi
    train_df = imputed_df[imputed_df[target_col].notna()].copy()

    X = train_df[numerical_features]
    y = train_df[target_col]

    # Encode variabel target menjadi numerik
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # Inisialisasi XGBoost Classifier
    classifier = xgb.XGBClassifier(
        objective='multi:softprob',
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42
    )

    # Sesuaikan jumlah lipatan (cv) agar tidak melebihi jumlah anggota kelas terkecil
    min_class_count = min(pd.Series(y_encoded).value_counts())
    cv_folds = max(2, min_class_count) if min_class_count > 1 else 2

    # Lakukan cross-validation dan simpan skor akurasi rata-rata
    cv_scores = cross_val_score(classifier, X, y_encoded, cv=cv_folds, scoring='accuracy')
    results[method_name] = cv_scores.mean()
    print(f"  - Akurasi dengan Imputasi {method_name} (cv={cv_folds}): {results[method_name]:.4f}")

print("-" * 50)
# ==============================================================================
# 5. VISUALISASI HASIL EVALUASI
# ==============================================================================
results_df = pd.DataFrame(list(results.items()), columns=['Method', 'Accuracy']).sort_values('Accuracy', ascending=False)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Method', y='Accuracy', data=results_df, palette='cividis')
plt.title(f'Akurasi XGBoost vs. Metode Imputasi Numerik\n(untuk memprediksi "{target_col}")', fontsize=16)
plt.xlabel('Metode Imputasi Fitur Numerik', fontsize=12)
plt.ylabel('Akurasi Rata-rata Cross-Validation', fontsize=12)
plt.ylim(0, 1.0)

# Menambahkan label nilai di atas setiap bar
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.4f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 9),
                     textcoords='offset points',
                     fontsize=11)

plt.tight_layout()
plt.savefig('imputation_method_comparison_xgboost.png')
plt.show()

print("Visualisasi 'imputation_method_comparison_xgboost.png' telah dibuat.")
print("-" * 50)

# ==============================================================================
# 6. IMPUTASI FINAL MENGGUNAKAN METODE TERBAIK
# ==============================================================================
best_method_name = max(results, key=results.get)
final_df = df.copy()
best_imputed_numerical_df = imputation_methods[best_method_name]

print(f"Metode terbaik adalah '{best_method_name}'. Melakukan imputasi final pada '{target_col}'...")

# Isi nilai numerik yang hilang di dataframe final dengan metode terbaik
final_df[numerical_features] = best_imputed_numerical_df[numerical_features]

# Pisahkan data untuk training model final
train_df_final = final_df[final_df[target_col].notna()]
impute_df_final = final_df[final_df[target_col].isna()]

if not impute_df_final.empty:
    X_train = train_df_final[numerical_features]
    y_train = train_df_final[target_col]
    X_to_impute = impute_df_final[numerical_features]

    le_final = LabelEncoder()
    y_train_encoded = le_final.fit_transform(y_train)

    # Latih classifier final pada semua data yang tersedia
    final_classifier = xgb.XGBClassifier(
        objective='multi:softprob',
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42
    )
    final_classifier.fit(X_train, y_train_encoded)

    # Prediksi nilai yang hilang
    predicted_labels_encoded = final_classifier.predict(X_to_impute)
    predicted_categories = le_final.inverse_transform(predicted_labels_encoded)

    # Isi nilai yang hilang di dataframe final
    final_df.loc[final_df[target_col].isna(), target_col] = predicted_categories
    print(f"Imputasi untuk '{target_col}' selesai.")
else:
    print(f"Tidak ada nilai yang hilang di '{target_col}' untuk diimputasi.")

# ==============================================================================
# 7. TAMPILKAN HASIL AKHIR
# ==============================================================================
print("-" * 50)
print(f"Contoh DataFrame setelah imputasi dengan metode '{best_method_name}':")
print(final_df.head(10))

print("\nJumlah missing value setelah imputasi final:")
print(final_df.isnull().sum())

# Simpan hasil akhir ke file CSV baru
final_df.to_csv('train_dataset_final_imputed_xgboost.csv', index=False)
print("\nDataset yang telah diimputasi disimpan sebagai 'train_dataset_final_imputed_xgboost.csv'")
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import lightgbm as lgb
import catboost as cb
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.decomposition import PCA
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# ==============================================================================
# 1. MEMBACA DATA
# ==============================================================================
# Gunakan data train yang SUDAH diimputasi dan di-encode
data_train = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/train_dataset_label_encoded_imputed.csv")
print("Data train yang telah diimputasi berhasil dimuat.")

# Muat data test dan submission
data_test = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/test_dataset.csv")
submission = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/Sample_Submission.csv")

# ==============================================================================
# 2. PRA-PEMROSESAN DATA TEST
# ==============================================================================
print("\nMemulai pra-pemrosesan pada data test...")
# Simpan ID untuk submission
test_ids = data_test['id']

# Hapus kolom yang tidak diperlukan
if 'school_name' in data_test.columns:
    data_test = data_test.drop(columns=['school_name', 'id'])

# Pisahkan kolom numerik dan kategorikal pada data test
num_cols_test = data_test.select_dtypes(include=['float64', 'int64']).columns
cat_cols_test = data_test.select_dtypes(include=['object', 'bool']).columns

# Imputasi sederhana (modus) untuk kolom kategorikal di data test
for col in cat_cols_test:
    mode_val = data_test[col].mode()[0]
    data_test[col].fillna(mode_val, inplace=True)

# Imputasi lanjutan untuk kolom numerik di data test
imputer = IterativeImputer(estimator=xgb.XGBRegressor(), max_iter=10, random_state=42)
data_test[num_cols_test] = imputer.fit_transform(data_test[num_cols_test])
print("Imputasi data test selesai.")

# ==============================================================================
# 3. FEATURE ENGINEERING
# ==============================================================================
# Definisikan kembali fungsi feature engineering
def feature_engineering(df):
    df = df.copy()
    # Fitur-fitur ini akan dibuat jika kolom aslinya ada
    if 'avg_test_score_percent' in df.columns and 'funding_per_student_usd' in df.columns:
        df['funding_efficiency'] = df['avg_test_score_percent'] / (df['funding_per_student_usd'] + 1)
    if 'student_teacher_ratio' in df.columns and 'percent_low_income' in df.columns:
        df['support_index'] = (1 / (df['student_teacher_ratio'] + 1)) * (1 - df['percent_low_income'] / 100)
    if 'internet_access_percent' in df.columns and 'percent_minority' in df.columns:
        df['equity_index'] = df['internet_access_percent'] / (1 + df['percent_minority'])
    return df

print("\nMenerapkan feature engineering...")
# Karena data train yang diimputasi tidak punya fitur interaksi, kita buat di sini
train_fe = feature_engineering(data_train)
test_fe = feature_engineering(data_test)

# ==============================================================================
# 4. ENCODING & SINKRONISASI KOLOM
# ==============================================================================
# Data train sudah di-encode. Kita hanya perlu encode data test.
# Gunakan OneHotEncoder pada data test
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
encoded_cols = [f"state_{c}" for c in data_test['state'].unique()] + \
               [f"school_type_{c}" for c in data_test['school_type'].unique()] + \
               [f"grade_level_{c}" for c in data_test['grade_level'].unique()]

# Encode kolom kategorikal pada data test
test_cat_encoded = pd.DataFrame(encoder.fit_transform(test_fe[cat_cols_test]),
                                columns=encoder.get_feature_names_out(cat_cols_test))

# Gabungkan fitur numerik & kategorik yang sudah di-encode untuk data test
test_final = pd.concat([
    test_fe.drop(columns=cat_cols_test).reset_index(drop=True),
    test_cat_encoded.reset_index(drop=True)
], axis=1)

# Menyiapkan data untuk model
target = 'dropout_rate_percent'
y = train_fe[target]
X = train_fe.drop(columns=[target])
X_test = test_final

# SINKRONISASI FINAL: Pastikan train dan test memiliki kolom yang persis sama
X_test = X_test.reindex(columns=X.columns, fill_value=0)
print("Data train dan test telah siap dan disinkronkan.")

# ==============================================================================
# 5. PCA (Principal Component Analysis)
# ==============================================================================
# Karena semua data sudah numerik, kita terapkan PCA ke seluruh fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

pca = PCA(n_components=0.95) # Mempertahankan 95% varians
X_pca = pca.fit_transform(X_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Konversi hasil PCA kembali ke DataFrame
X_pca = pd.DataFrame(X_pca)
X_test_pca = pd.DataFrame(X_test_pca)
print(f"\nPCA diterapkan, jumlah komponen utama: {X_pca.shape[1]}")

# ==============================================================================
# 6. TRAINING & EVALUASI MODEL
# ==============================================================================
def model_training(X, y, X_test, model_type):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    # Ubah kolom menjadi string untuk kompatibilitas
    X.columns = X.columns.astype(str)
    X_test.columns = X_test.columns.astype(str)

    if model_type == 'xgboost':
        model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='mae', random_state=42)
        param_grid = {'max_depth': [6, 8], 'learning_rate': [0.05, 0.1], 'n_estimators': [200, 300]}
    elif model_type == 'catboost':
        model = cb.CatBoostRegressor(loss_function='MAE', verbose=0, random_seed=42)
        param_grid = {'depth': [6, 8], 'learning_rate': [0.05, 0.1], 'iterations': [200, 300]}
    elif model_type == 'lightgbm':
        model = lgb.LGBMRegressor(objective='regression_l1', metric='mae', random_state=42)
        param_grid = {'max_depth': [6, 8], 'learning_rate': [0.05, 0.1], 'n_estimators': [200, 300]}

    grid_search = GridSearchCV(model, param_grid, cv=kf, scoring='neg_mean_absolute_error', n_jobs=-1)
    grid_search.fit(X, y)
    best_model = grid_search.best_estimator_
    print(f"\nParameter terbaik untuk {model_type}: {grid_search.best_params_}")

    # Latih kembali dengan model terbaik pada cross-validation
    oof_preds = np.zeros(len(X))
    test_preds = np.zeros(len(X_test))
    for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        best_model.fit(X_train, y_train)
        oof_preds[val_idx] = best_model.predict(X_val)
        test_preds += best_model.predict(X_test) / kf.n_splits

    mae_score = mean_absolute_error(y, oof_preds)
    print(f"Rata-rata MAE ({model_type}): {mae_score:.4f}")
    return mae_score, test_preds

# Melakukan pelatihan untuk setiap model
xgb_mae, xgb_test_preds = model_training(X_pca, y, X_test_pca, 'xgboost')
catboost_mae, catboost_test_preds = model_training(X_pca, y, X_test_pca, 'catboost')
lightgbm_mae, lightgbm_test_preds = model_training(X_pca, y, X_test_pca, 'lightgbm')

# ==============================================================================
# 7. SUBMISSION
# ==============================================================================
mae_scores = {'XGBoost': xgb_mae, 'CatBoost': catboost_mae, 'LightGBM': lightgbm_mae}
preds = {'XGBoost': xgb_test_preds, 'CatBoost': catboost_test_preds, 'LightGBM': lightgbm_test_preds}

# Pilih model dengan MAE terendah
best_model_name = min(mae_scores, key=mae_scores.get)
best_model_preds = preds[best_model_name]

print(f"\nModel terbaik adalah {best_model_name} dengan MAE: {mae_scores[best_model_name]:.4f}")

submission['dropout_rate_percent'] = best_model_preds
submission.to_csv("submission_final.csv", index=False)

print("File submission 'submission_final.csv' telah berhasil dibuat.")
print(submission.head())
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import lightgbm as lgb
import catboost as cb
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.decomposition import PCA
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler

# ==============================================================================
# 1. MEMBACA DATA
# ==============================================================================
# Gunakan data train yang SUDAH diimputasi dan di-encode dengan LabelEncoder
data_train = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/train_dataset_label_encoded_imputed.csv")
print("Data train yang telah diimputasi berhasil dimuat.")

# Muat data test dan submission
data_test = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/test_dataset.csv")
submission = pd.read_csv(r"C:/Uner/Lomba/Shine/rumpun-penalaran-2025-data-science-competition/Sample_Submission.csv")

# ==============================================================================
# 2. PRA-PEMROSESAN DATA TEST
# ==============================================================================
print("\nMemulai pra-pemrosesan pada data test...")
test_ids = data_test['id']
if 'school_name' in data_test.columns:
    data_test = data_test.drop(columns=['school_name', 'id'])

num_cols_test = data_test.select_dtypes(include=['float64', 'int64']).columns
cat_cols_test = data_test.select_dtypes(include=['object', 'bool']).columns

# Imputasi modus untuk kategorikal
for col in cat_cols_test:
    mode_val = data_test[col].mode()[0]
    data_test[col].fillna(mode_val, inplace=True)

# Imputasi Iterative untuk numerik
imputer = IterativeImputer(estimator=xgb.XGBRegressor(), max_iter=10, random_state=42)
data_test[num_cols_test] = imputer.fit_transform(data_test[num_cols_test])
print("Imputasi data test selesai.")

# ==============================================================================
# 3. FEATURE ENGINEERING
# ==============================================================================
def feature_engineering(df):
    df = df.copy()
    if 'avg_test_score_percent' in df.columns and 'funding_per_student_usd' in df.columns:
        df['funding_efficiency'] = df['avg_test_score_percent'] / (df['funding_per_student_usd'] + 1)
    if 'student_teacher_ratio' in df.columns and 'percent_low_income' in df.columns:
        df['support_index'] = (1 / (df['student_teacher_ratio'] + 1)) * (1 - df['percent_low_income'] / 100)
    if 'internet_access_percent' in df.columns and 'percent_minority' in df.columns:
        df['equity_index'] = df['internet_access_percent'] / (1 + df['percent_minority'])
    return df

print("\nMenerapkan feature engineering...")
train_fe = feature_engineering(data_train)
test_fe = feature_engineering(data_test)

# ==============================================================================
# 4. ENCODING & SINKRONISASI KOLOM
# ==============================================================================
# KONSISTENSI: Gunakan LabelEncoder pada data test, sama seperti data train
print("Melakukan Label Encoding pada data test...")
for col in cat_cols_test:
    # Buat encoder baru untuk setiap kolom
    le = LabelEncoder()
    # Fit encoder pada kategori yang ada di data test
    le.fit(test_fe[col].unique())
    # Transform data test
    test_fe[col] = le.transform(test_fe[col])

target = 'dropout_rate_percent'
y = train_fe[target]
X = train_fe.drop(columns=[target])
X_test = test_fe

# SINKRONISASI FINAL: Pastikan train dan test memiliki kolom yang persis sama
X_test = X_test.reindex(columns=X.columns, fill_value=0)
print("Data train dan test telah siap dan disinkronkan.")

# ==============================================================================
# 5. PCA (Principal Component Analysis)
# ==============================================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

pca = PCA(n_components=0.95) # Mempertahankan 95% varians
X_pca = pca.fit_transform(X_scaled)
X_test_pca = pca.transform(X_test_scaled)

X_pca = pd.DataFrame(X_pca)
X_test_pca = pd.DataFrame(X_test_pca)
print(f"\nPCA diterapkan, jumlah komponen utama: {X_pca.shape[1]}")

# ==============================================================================
# 6. TRAINING & EVALUASI MODEL DENGAN TUNING MAKSIMAL
# ==============================================================================
def model_training(X, y, X_test, model_type):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    X.columns, X_test.columns = X.columns.astype(str), X_test.columns.astype(str)

    # --- GRID PARAMETER YANG DIOPTIMALKAN ---
    if model_type == 'xgboost':
        model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='mae', random_state=42, n_jobs=-1)
        param_grid = {
            'max_depth': [5, 7, 9],
            'learning_rate': [0.01, 0.05, 0.1],
            'n_estimators': [200, 400, 600],
            'subsample': [0.7, 0.9],
            'colsample_bytree': [0.7, 0.9]
        }
    elif model_type == 'catboost':
        model = cb.CatBoostRegressor(loss_function='MAE', verbose=0, random_seed=42)
        param_grid = {
            'depth': [6, 8, 10],
            'learning_rate': [0.03, 0.05, 0.1],
            'iterations': [300, 500, 700],
            'l2_leaf_reg': [1, 3, 5] # L2 Regularization
        }
    elif model_type == 'lightgbm':
        model = lgb.LGBMRegressor(objective='regression_l1', metric='mae', random_state=42, n_jobs=-1)
        param_grid = {
            'max_depth': [5, 7, 9],
            'learning_rate': [0.01, 0.05, 0.1],
            'n_estimators': [200, 400, 600],
            'subsample': [0.7, 0.9],
            'colsample_bytree': [0.7, 0.9]
        }
    
    # Gunakan GridSearchCV dengan grid yang sudah dioptimalkan
    grid_search = GridSearchCV(model, param_grid, cv=kf, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=1)
    grid_search.fit(X, y)
    best_model = grid_search.best_estimator_
    print(f"\nParameter terbaik untuk {model_type}: {grid_search.best_params_}")

    oof_preds = np.zeros(len(X))
    test_preds = np.zeros(len(X_test))
    for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        best_model.fit(X_train, y_train)
        oof_preds[val_idx] = best_model.predict(X_val)
        test_preds += best_model.predict(X_test) / kf.n_splits

    mae_score = mean_absolute_error(y, oof_preds)
    print(f"Rata-rata MAE ({model_type}): {mae_score:.4f}")
    return mae_score, test_preds

# Melakukan pelatihan untuk setiap model
xgb_mae, xgb_test_preds = model_training(X_pca, y, X_test_pca, 'xgboost')
catboost_mae, catboost_test_preds = model_training(X_pca, y, X_test_pca, 'catboost')
lightgbm_mae, lightgbm_test_preds = model_training(X_pca, y, X_test_pca, 'lightgbm')

# ==============================================================================
# 7. SUBMISSION
# ==============================================================================
mae_scores = {'XGBoost': xgb_mae, 'CatBoost': catboost_mae, 'LightGBM': lightgbm_mae}
preds = {'XGBoost': xgb_test_preds, 'CatBoost': catboost_test_preds, 'LightGBM': lightgbm_test_preds}

best_model_name = min(mae_scores, key=mae_scores.get)
best_model_preds = preds[best_model_name]

print(f"\nModel terbaik adalah {best_model_name} dengan MAE: {mae_scores[best_model_name]:.4f}")

submission['dropout_rate_percent'] = best_model_preds
submission.to_csv("submission_final_optimized.csv", index=False)

print("File submission 'submission_final_optimized.csv' telah berhasil dibuat.")
print(submission.head())
```
